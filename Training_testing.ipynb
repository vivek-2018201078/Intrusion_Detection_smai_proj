```python
import pandas as pd
import numpy as np
import time
import math
```

```python
def distance_from_center(sample, center) :
    s_cont = sample[0:35]
    s_cat = sample[35:41]
    c_cont = center[0:35]
    c_cat = center[34:41]
#     print("Sample continous : ", s_cont )
#     print("Center continous : ", c_cont )
    x = np.subtract(s_cont,c_cont)
#     print(x, "after substraction")
    d = np.sum(np.square(x))
#     print(d, " before adding categorical distance")
    for i in range(6) :
        if s_cat[i] != c_cat[i] :
            d += 1
    return d
```

```python
def training(filename, W) :
    normalised_train = np.array(pd.read_csv(filename))
    normalised_train = normalised_train[:,0:41]
    centers = []
    clusters = {}
    sum_total = []
#     W = 5
    start = time.time()
    max_d = 0
    for i in range(len(normalised_train)) :
        if i%1000 == 999 :
            print("For ", i//1000, " numbered block, time taken : ", time.time() - start)
            start = time.time() 
            print("Length of centers : ", len(centers))
        if len(centers) == 0 :
            centers.append(normalised_train[i])
            clusters[0] = 1
#             print(centers[0])
        else :
            ind = 0 
            mindist = 100000000
#             print(len(centers))
            for j in range(len(centers)) :
                d = distance_from_center(normalised_train[i], centers[j])
                if d > max_d : 
                    max_d = d
                if mindist > d :
                    mindist = d
                    ind = j
            if mindist > W :
                centers.append(normalised_train[i])
                clusters[len(centers) - 1] = 1
            else : 
                clusters[ind] += 1
                mod_center = centers[ind]
    print("Maximum possible distance : ", max_d)
    return centers, clusters
#     N = 0.15
```

```python
def normal_clusters_identification(clusters, N) :
    total_normal = math.ceil(len(clusters) * N)
    cluster_number = list(clusters.keys())
    cluster_count = list(clusters.values())
    normal_clusters = []
    for i in range(total_normal) :
        ind = cluster_count.index(max(cluster_count))
        normal_clusters.append(ind)
        cluster_count[ind]= 0
        
    return normal_clusters
```

```python
def testing(filename, centers, normal_clusters) :
    n_n = 0
    n_a = 0
    a_n = 0
    a_a = 0
#     count = 0
    normalised_test = np.array(pd.read_csv(filename))
    normalised_test = normalised_test[:, 0:42]
    start = time.time()
    for i in range(len(normalised_test)) :
        if i%1000 == 999 :
            print("For ", i//1000, " numbered block, time taken : ", time.time() - start)
            start = time.time()
#             count += 1
#         if count == 10 :
#             break
        mindistance = 100000000
        ind = 0
        for j in range(len(centers)) :
            d = distance_from_center(normalised_test[i, 0:41], centers[j])
            if d < mindistance :
                mindistance = d
                ind = j
        if ind in normal_clusters :
            if normalised_test[i, -1] == "normal." :
                n_n += 1
            else :
                a_n += 1
        else :
            if normalised_test[i, -1] == "normal." :
                n_a += 1
            else :
                a_a += 1
    return n_n, n_a, a_n, a_a 
```

```python
W = 40
centers, clusters = training("../input/normalised_block_10.csv", W )#filename, W
```

```python
N = 0.15
normal_clusters = normal_clusters_identification(clusters, N)
```
